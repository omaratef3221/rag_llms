```python
from RAG import RAG_LLM

MyObj = RAG_LLM(pdf_path = "Supervised Learning Book .pdf",
                  model_path = "mistralai/Mistral-7B-v0.1",
                  with_hf_api = False,
                  device= 'cuda',
                  hf_api_key = "")
```

```python
pages = MyObj.load_pdf(page_range = [6, -2])
chunks = MyObj.get_chunks(pages)
```

```python
db_store, Embeddings = MyObj.create_faiss_embeddings(chunks)
```

```python
prompt_str = """
                You are an Artificial intelligence Teacher, helping AI students and answering their questions.
                If the context is not related to AI, say "I don't know"

                Based on this Context:

                "{context}"

                \n\n

                What is the answer of this question:
                "{question}"

                Answer:
                """
prompt = MyObj.get_prompt(prompt = prompt_str)
```

```python
llm_model = MyObj.get_model(quantization = True, max_length = 128)
```

```python
query1 = "what is the maximum likelihood equation?"
answer1 = MyObj.get_response(query1, Embeddings, db_store, prompt, llm_model)
print(answer1["query"])
print("="*20)
print(answer1["result"])
print("="*20)
print(answer1["source_documents"])
```

```python
query2 = "what is rbf and how does it work?"
answer2 = MyObj.get_response(query2, Embeddings, db_store, prompt, llm_model)
print(answer2["query"])
print("="*20)
print(answer2["result"])
print("="*20)
print(answer2["source_documents"])
```

```python
import pandas as pd
from langchain.vectorstores.chroma import Chroma


docs_vectorstore = Chroma(
    collection_name="docs_store",
    embedding_function=Embeddings,
    persist_directory="docs-db",
)


response = docs_vectorstore.get(include=["metadatas", "documents", "embeddings"])
df = pd.DataFrame(
    {
        "id": response["ids"],
        "source": [metadata.get("source") for metadata in response["metadatas"]],
        "page": [metadata.get("page", -1) for metadata in response["metadatas"]],
        "document": response["documents"],
        "embedding": response["embeddings"],
    }
)
df["contains_answer"] = df["document"].apply(lambda x: "Eichler" in x)
df["contains_answer"].to_numpy().nonzero()


import hashlib
import json
from langchain_core.documents import Document

def stable_hash(doc: Document) -> str:
    """
    Stable hash document based on its metadata.
    """
    return hashlib.sha1(json.dumps(doc.metadata, sort_keys=True).encode()).hexdigest()

split_ids = list(map(stable_hash, chunks))

docs_vectorstore.add_documents(chunks, ids=split_ids)
docs_vectorstore.persist()
```
